```python
import numpy as np
import pandas as pd
from numpy.random import randn
```

```python
np.random.seed(101)
```

```python
# create DataFrame df
df = pd.DataFrame(randn(5, 4), ['A', 'B', 'C', 'D', 'E'], ['W', 'X', 'Y', 'Z'])
df
```

```python
# create boolean DataFrame for values greater than zero
booldf = df > 0
booldf
```

```python
# pass boolean DataFrame into original DataFrame
df[booldf]
```

```python
# this is the same conditional selection as above, but simpler (no need for booldf)
df[df > 0]
```

```python
# view boolean Series where df['W'] is greater than 0
df['W'] > 0
```

```python
# view df['W'] (creating a boolean Series did not change the original)
df['W']
```

```python
# view df where df['W'] is greater than 0
df[df['W'] > 0]
```

```python
# view df where df['Z'] is less than 0
df[df['Z'] < 0]
```

```python
# create new DataFrame using comparison operator to eliminate rows where df['W'] is greater than 0
resultdf = df[df['W'] > 0]
```

```python
# view resultdf
resultdf
```

```python
# view Series resultdf['X']
resultdf['X']
```

```python
# show columns X and Y when df['W'] > 0
df[df['W'] > 0][['X', 'Y']]
```

```python
# below code is equivalent to df[df['W'] > 0][['X', 'Y']], seen above
boolser = df['W'] > 0
result = df[boolser]
mycols = ['X', 'Y']
result[mycols]
```

```python
# view df where df['W'] is greater than 0 AND df['Y'] is greater than 1
df[(df['W'] > 0) & (df['Y'] > 1)]
```

```python
# view df where df['W'] is greater than 0 OR df['Y'] is greater than 1
df[(df['W'] > 0) | (df['Y'] > 1)]
```

```python
# view df (conditional selection does not automatically change the DataFrame, so it remains in its original state)
df
```

```python
# move current indices to new column and revert to default numerical indexing (does not occur in-place unless specified)
df.reset_index()
```

```python
# create list for new indices
newind = 'CA NY WY OR CO'.split()
```

```python
# check list of new indices
newind
```

```python
# create new 'States' column using newind as data (adding new column will change original DataFrame)
df['States'] = newind
df
```

```python
# set 'States' column to be new index (does not occur in-place unless specified)
df.set_index('States')
```

```python
# view df (the 'States' column is no longer the index because set_index('States') was not called in-place)
df
```
