```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# allows plots to be displayed in jupyter notebook
%matplotlib inline
```

```python
# create data to plot
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
# view array x
x
```

```python
# view array y
y
```

# Scripting Methods for Creating Plots

```python
# create simple plot
plt.plot(x, y)
```

```python
# continue building on previous plot by adding axis labels
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# plt.subplot(nrows, ncolumns, plot number)

# create two axes and plot on first axes
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.title('First Axes')

# plot on second axes
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
plt.title('Second Axes')
```

# Object-Oriented Methods for Creating Plots

```python
# create figure object
fig = plt.figure()

# add_axes([left, bottom, width, height]) - creates axes object
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# plot on axes
axes.plot(x, y)

# label axes
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
# create figure object (blank canvas)
fig = plt.figure()

# create axes and set positions
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])


# plot data on axes
axes1.plot(x, y)
axes2.plot(y, x)

# label axes
axes2.set_title('Smaller Plot (axes2)')
axes1.set_title('Larger Plot (axes1)')
```

```python
# create figure and add multiple axes  (two columns, one row)
fig, axes = plt.subplots(nrows = 1, ncols = 2)

# since plt.suplots() creates an array of axes, you can plot on each as you iterate through them
for current_ax in axes:
    current_ax.plot(x, y)
    
# set titles
axes[0].set_title('First Plot')
axes[1].set_title('Second Plot')
    
plt.tight_layout()
```

```python
# create same fig and axes as above
fig, axes = plt.subplots(nrows = 1, ncols = 2)

# plot via index of axes instead of using loop
axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')


plt.tight_layout()
```

# Figure Sizes, Saving Plots, and DPI

```python
# use tuple as argument to specify width and height of figure (can also specify resolution using dpi argument)
fig = plt.figure(figsize = (8, 2))

# create new axes and plot data
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
# create another figure with two axes (two rows, one column) and 5x5 size
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (5, 5))

# plot on axes
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

```python
# save the figure, specifying dpi (resolution)
fig.savefig('myFig.png', dpi = 100)
```

```python
# create new fig
fig = plt.figure()

# add axes
ax = fig.add_axes([0, 0, 1, 1])

# plot data
ax.plot(x, y)

# set title
ax.set_title('Title')

# set axis labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
```

# Using Legends

```python
# create figure
fig = plt.figure()

# create axes
ax = fig.add_axes([0, 0, 1, 1])

# plot different data on same axes. add label argument to be used in legend
ax.plot(x, x ** 2, label = 'x^2')
ax.plot(x, x ** 3, label = 'x^3')

# create legend
ax.legend(loc = 0)
```
